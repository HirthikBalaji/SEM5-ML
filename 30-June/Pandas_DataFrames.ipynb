### Program 1: Creating a DataFrame
We create a simple pandas DataFrame using a python dictionary.


```python
import pandas as pd
data = {
    'Name': ['Rahul', 'Priya', 'Amit'],
    'Age': [20, 21, 22],
    'Marks': [85, 90, 88]
}
df = pd.DataFrame(data)
print(df)
```

### Program 2: Displaying DataFrame Information
Displays columns, data types, non-null counts, and memory usage of the DataFrame.


```python
print(df.info())
```

### Program 3: Displaying First and Last Rows
Reads the `students.csv` file and displays the first 5 and last 5 rows.


```python
df_csv = pd.read_csv("students.csv")
print("First 5 Rows:")
print(df_csv.head())
print("\nLast 5 Rows:")
print(df_csv.tail())
```

### Program 4: Selecting a Single Column
Demonstrates how to select a single column from a DataFrame.


```python
print(df['Name'])
```

### Program 5: Selecting Multiple Columns
Demonstrates how to select multiple columns from a DataFrame.


```python
print(df[['Name', 'Marks']])
```

### Program 6: Selecting Rows Using loc
Demonstrates row selection by index label using `loc`.


```python
print(df.loc[1])
```

### Program 7: Filtering Data
Filters the DataFrame to find students with Marks greater than 80.


```python
result = df[df['Marks'] > 80]
print(result)
```

### Program 8: Adding a New Column
Adds a new 'Grade' column to the DataFrame.


```python
df['Grade'] = ['A', 'A', 'B']
print(df)
```

### Program 9: Updating Values
Modifies a specific cell value (row 0, column 'Marks') in the DataFrame.


```python
df.loc[0, 'Marks'] = 95
print(df)
```

### Program 10: Deleting a Column
Removes the 'Age' column from the DataFrame.


```python
df.drop('Age', axis=1, errors='ignore', inplace=True)
print(df)
```

### Program 11: Descriptive Statistics
Computes summary statistics for numerical columns in the DataFrame.


```python
data_stats = {
    'Marks': [85, 90, 88, 75, 95]
}
df_stats = pd.DataFrame(data_stats)
print(df_stats.describe())
```

### Program 12: Handling Missing Values
Fills missing values in a dataset with the column mean.


```python
data_nan = {
    'Marks': [85, None, 90, None, 88]
}
df_nan = pd.DataFrame(data_nan)
print("Before Filling:")
print(df_nan)
df_nan.fillna(df_nan['Marks'].mean(), inplace=True)
print("\nAfter Filling:")
print(df_nan)
```

### Program 13: Sorting Data
Sorts the DataFrame by 'Marks' in descending order.


```python
data_sort = {
    'Name': ['Rahul', 'Priya', 'Amit'],
    'Marks': [85, 90, 88]
}
df_sort = pd.DataFrame(data_sort)
print(df_sort.sort_values(by='Marks', ascending=False))
```

### Program 14: Saving DataFrame to CSV
Saves a DataFrame to a CSV file.


```python
data_save = {
    'Name': ['Rahul', 'Priya', 'Amit'],
    'Marks': [85, 90, 88]
}
df_save = pd.DataFrame(data_save)
df_save.to_csv("output.csv", index=False)
print("File Saved Successfully")
```
