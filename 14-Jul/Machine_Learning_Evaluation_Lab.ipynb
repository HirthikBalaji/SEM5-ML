## Experiment 1: Disease Prediction Binary Classification

### Aim
Evaluate a disease prediction system's performance by calculating precision, recall, F1-score, and displaying the confusion matrix.

### Problem Statement
Given the actual and predicted disease diagnoses (1 for disease present, 0 for healthy):
- **Actual:** `[1, 1, 1, 0, 0, 0, 1, 0, 1, 0, 1, 0]`
- **Predicted:** `[1, 0, 1, 0, 0, 1, 1, 0, 0, 0, 1, 0]`

### Mathematical Formulations

#### 1. Confusion Matrix
A tabular layout that visualizes the performance of a supervised learning algorithm:
$$
\text{Confusion Matrix} = \begin{pmatrix} \text{TN} & \text{FP} \\ \text{FN} & \text{TP} \end{pmatrix}
$$
Where:
- **True Positives (TP):** Actual 1, Predicted 1
- **False Positives (FP):** Actual 0, Predicted 1 (Type I error)
- **True Negatives (TN):** Actual 0, Predicted 0
- **False Negatives (FN):** Actual 1, Predicted 0 (Type II error)

#### 2. Precision
The proportion of positive identifications that were actually correct:
$$
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
$$

#### 3. Recall (Sensitivity / True Positive Rate)
The proportion of actual positives that were identified correctly:
$$
\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$

#### 4. F1-Score
The harmonic mean of precision and recall:
$$
\text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} = \frac{2\text{TP}}{2\text{TP} + \text{FP} + \text{FN}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# 1. Define dataset
actual_1 = np.array([1, 1, 1, 0, 0, 0, 1, 0, 1, 0, 1, 0])
predicted_1 = np.array([1, 0, 1, 0, 0, 1, 1, 0, 0, 0, 1, 0])

# 2. Calculate metrics using scikit-learn
precision = precision_score(actual_1, predicted_1)
recall = recall_score(actual_1, predicted_1)
f1 = f1_score(actual_1, predicted_1)
cm = confusion_matrix(actual_1, predicted_1)

# 3. Print results
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"\nConfusion Matrix:\n{cm}")

# 4. Plot Confusion Matrix Heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Negative (0)', 'Positive (1)'],
            yticklabels=['Negative (0)', 'Positive (1)'])
plt.title('Confusion Matrix - Experiment 1')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## Experiment 2: Spam Detection Classifier Evaluation

### Aim
Evaluate a spam detection classifier by converting categorical text labels to binary values and calculating precision, recall, and F1-score.

### Problem Statement
Given labels:
- **Actual:** `['Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam']`
- **Predicted:** `['Spam', 'Spam', 'Not Spam', 'Not Spam', 'Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam']`

### Strategy
1. Map `'Spam'` to positive class `1` and `'Not Spam'` to negative class `0`.
2. Perform binary metrics calculations.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# 1. Define categorical dataset
actual_2_str = ['Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam']
predicted_2_str = ['Spam', 'Spam', 'Not Spam', 'Not Spam', 'Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam']

# 2. Convert to binary values
mapping = {'Spam': 1, 'Not Spam': 0}
actual_2 = np.array([mapping[item] for item in actual_2_str])
predicted_2 = np.array([mapping[item] for item in predicted_2_str])

print(f"Binary Mapping: Spam -> 1, Not Spam -> 0")
print(f"Actual Binary: {list(actual_2)}")
print(f"Predicted Binary: {list(predicted_2)}\n")

# 3. Calculate metrics
precision = precision_score(actual_2, predicted_2)
recall = recall_score(actual_2, predicted_2)
f1 = f1_score(actual_2, predicted_2)
cm = confusion_matrix(actual_2, predicted_2)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"\nConfusion Matrix:\n{cm}")

# 4. Plot Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', cbar=False,
            xticklabels=['Not Spam (0)', 'Spam (1)'],
            yticklabels=['Not Spam (0)', 'Spam (1)'])
plt.title('Confusion Matrix - Experiment 2 (Spam Classifier)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## Experiment 3: Iris Dataset Multi-class Classification & Decision Tree

### Aim
Generate a complete classification report for a Decision Tree classifier trained on the standard Iris dataset.

### Problem Statement
1. Load the Iris dataset.
2. Split the dataset into training and testing sets (70-30 split).
3. Train a Decision Tree classifier.
4. Predict the test data.
5. Calculate Precision, Recall, F1-score, and Accuracy using Scikit-learn.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# 1. Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# 2. Split dataset into training and testing sets (70/30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Train a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42, max_depth=3)
clf.fit(X_train, y_train)

# 4. Predict the test data
y_pred = clf.predict(X_test)

# 5. Calculate Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Dataset split: 70% Training, 30% Testing\n")
print(f"Overall Accuracy: {accuracy:.4f}")
print(f"Weighted Precision: {precision:.4f}")
print(f"Weighted Recall: {recall:.4f}")
print(f"Weighted F1-score: {f1:.4f}\n")

# 6. Generate Complete Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# 7. Plot the Trained Decision Tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title('Decision Tree Classifier Trained on Iris Dataset')
plt.show()
```

## Experiment 4: Temperature Prediction Regression Metrics

### Aim
Calculate regression error metrics for a temperature prediction model.

### Problem Statement
Given the actual and predicted daily temperatures (°C):
- **Actual:** `[30, 32, 31, 35, 36, 34, 33]`
- **Predicted:** `[29, 33, 30, 34, 37, 35, 32]`

### Mathematical Formulations

#### 1. Mean Absolute Error (MAE)
The average of the absolute differences between predicted values and actual values:
$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

#### 2. Mean Squared Error (MSE)
The average of the squared differences between predicted values and actual values (penalizes larger errors):
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

#### 3. Root Mean Squared Error (RMSE)
The square root of the Mean Squared Error (restores the unit of scale to match original data):
$$
\text{RMSE} = \sqrt{\text{MSE}} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Define temperature dataset
actual_4 = np.array([30, 32, 31, 35, 36, 34, 33])
predicted_4 = np.array([29, 33, 30, 34, 37, 35, 32])

# 2. Calculate regression metrics
mae = mean_absolute_error(actual_4, predicted_4)
mse = mean_squared_error(actual_4, predicted_4)
rmse = np.sqrt(mse)

# 3. Print results
print(f"Actual Temperatures: {list(actual_4)}")
print(f"Predicted Temperatures: {list(predicted_4)}\n")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
```

## Experiment 5: Student Exam Marks Regression Evaluation & Interpretation

### Aim
Evaluate the performance of a regression model predicting students' final examination marks using MAE, MSE, and RMSE, and interpret the obtained error values.

### Problem Statement
A regression model predicts students' final examination marks:
- **Actual Marks:** `[78, 85, 90, 72, 88, 95, 80]`
- **Predicted Marks:** `[80, 82, 89, 75, 86, 93, 79]`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Define marks dataset
actual_5 = np.array([78, 85, 90, 72, 88, 95, 80])
predicted_5 = np.array([80, 82, 89, 75, 86, 93, 79])

# 2. Calculate regression metrics
mae = mean_absolute_error(actual_5, predicted_5)
mse = mean_squared_error(actual_5, predicted_5)
rmse = np.sqrt(mse)

# 3. Print results
print(f"Actual Marks: {list(actual_5)}")
print(f"Predicted Marks: {list(predicted_5)}\n")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# 4. Plot actual vs predicted values
plt.figure(figsize=(7, 5))
indices = np.arange(len(actual_5))
width = 0.35
plt.bar(indices - width/2, actual_5, width, label='Actual Marks', color='royalblue')
plt.bar(indices + width/2, predicted_5, width, label='Predicted Marks', color='coral')
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.title('Actual vs Predicted Marks (Experiment 5)')
plt.xticks(indices)
plt.legend()
plt.show()
```

### Interpretation of Obtained Error Values (Experiment 5)

1. **Mean Absolute Error (MAE) = 2.0000**
   - **Meaning:** On average, the model's predicted exam mark deviates from the actual exam mark by **exactly 2.0 marks**.
   - **Property:** MAE provides a linear penalty where all errors are weighted equally. This is a robust indicator of the average expected prediction error magnitude.

2. **Mean Squared Error (MSE) = 4.5714**
   - **Meaning:** The average of the squared prediction errors is **4.5714 marks²**.
   - **Property:** By squaring the errors prior to averaging, MSE penalizes larger prediction discrepancies much more heavily than smaller ones. Because the units are squared, it is not directly interpretable in the context of student marks, but serves as a useful optimization objective.

3. **Root Mean Squared Error (RMSE) = 2.1381**
   - **Meaning:** The standard deviation of the prediction residuals is approximately **2.1381 marks**.
   - **Property:** RMSE is in the same units as the original target variable (marks), making it highly intuitive. Since $\text{RMSE} > \text{MAE}$ ($2.1381 > 2.0$), it indicates that there is some variance in the error magnitudes (namely, the presence of slightly larger errors of $\pm 3$ marks for indices 1 and 3, which pull the RMSE upward due to the squaring effect).

### Conclusion
Overall, the regression model is highly accurate, with an average prediction deviation of only 2.0 to 2.14 marks on a scale of exam scores ranging from 72 to 95.
